```python
import math

def is_prime(n):
    if n <= 1:
        return False
    for m in range(2, int(math.sqrt(n)) + 1):
        if n % m == 0:
            return False
    return True

# Test the program with the given numbers
numbers_to_test = [8, 105, 177, 51, 5, 47, 199, 967]

for num in numbers_to_test:
    if is_prime(num):
        print(f"{num} is prime")
    else:
        print(f"{num} is not prime")

# Test all numbers from 2 to 100 systematically
for num in range(2, 101):
    if is_prime(num):
        print(f"{num} is prime")
    else:
        print(f"{num} is not prime")
```
